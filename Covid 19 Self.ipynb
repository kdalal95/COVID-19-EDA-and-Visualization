```python
df= pd.read_excel('Covid cases in India.xlsx')
```

```python
df
```

```python
df.drop(['S. No.'], axis= 1)
```

```python
df['Total Cases'] = df['Total Confirmed cases (Indian National)'] + df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
total_cases_intotal= df['Total Cases'].sum()
print('Total cases in India are :', total_cases_intotal)
```

```python
df['Total Active Cases']= df['Total Cases'] - (df['Cured'] + df['Death'])
```

```python
df
```

```python
df.style.background_gradient(cmap='Reds')
```

```python
Total_Highest_Cases= df.groupby('Name of State / UT')['Total Cases','Total Active Cases'].sum().sort_values('Total Cases', ascending=False)
```

```python
fig= make_subplots(rows=2, cols=2,
                   specs=[[{'secondary_y': True}, {'secondary_y': True}], [{'secondary_y': True}, {'secondary_y': True}]],
subplot_titles=('Korea', 'Italy', 'India', 'Wuhan'))


fig.add_trace(go.Bar(x=db_korea['Date'], y=db_korea['Total Cases'], 
                     marker= dict(color= db_korea['Total Cases'], coloraxis= 'coloraxis')),1,1)

fig.add_trace(go.Bar(x=db_italy['Date'], y=db_italy['Total Cases'], 
                     marker= dict(color= db_italy['Total Cases'], coloraxis= 'coloraxis')),1,2)

fig.add_trace(go.Bar(x=db_india['Date'], y=db_india['Total Cases'], 
                     marker= dict(color= db_india['Total Cases'], coloraxis= 'coloraxis')),2,1)

fig.add_trace(go.Bar(x=db_wuhan['Date'], y=db_wuhan['Total Cases'], 
                     marker= dict(color= db_wuhan['Total Cases'], coloraxis= 'coloraxis')),2,2)


fig.update_layout(coloraxis= dict(colorscale= 'Bluered_r'), showlegend= False, title_text= 'Total cases in 4 Counteries')

fig.update_layout(plot_bgcolor= 'rgb(230,230,230)')
```

```python
df = pd.read_csv('covid_19_data.csv', parse_dates=['Last Update'])
```

```python
df.rename(columns={'ObservationDate': 'Date', 'Country/Region': 'Country' }, inplace= True)
```

```python
df.head()
```

```python
df.query('Country== "India"')
```

```python
df.groupby('Date').sum()
```

```python
confirmed= df.groupby('Date').sum()['Confirmed'].reset_index()
death= df.groupby('Date').sum()['Deaths'].reset_index()
rec= df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
fig= go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'], y=confirmed['Confirmed'], mode= 'lines+markers', 
                         name= 'Confirmed', line= dict(color= 'blue')))

fig.add_trace(go.Scatter(x=death['Date'], y=death['Deaths'], mode= 'lines+markers', 
                         name= 'Deaths', line= dict(color= 'red')))

fig.add_trace(go.Scatter(x=rec['Date'], y=rec['Recovered'], mode= 'lines+markers', 
                         name= 'Recovered', line= dict(color= 'green')))
```

```python
df_confirmed= pd.read_csv('time_series_covid_19_confirmed.csv')
```

```python
df_confirmed.rename(columns={'Country/Region': 'Country'}, inplace= True)
```

```python
df_latlong= pd.merge(df, df_confirmed, on= ['Country', 'Province/State'])
```

```python
df_latlong.sort_values('Date', inplace= True)
```

```python
fig=px.density_mapbox(df_latlong, lat= 'Lat', lon= 'Long', hover_name= 'Country', 
                      hover_data=['Confirmed', 'Deaths', 'Recovered'], animation_frame= 'Date', 
                      color_continuous_scale='Portland', radius= 7,zoom=0, height= 1000)

fig.update_layout(title= 'Woldwide Covid-19 Cases')

fig.update_layout(mapbox_style='open-street-map', mapbox_center_lon=0)

fig.update_layout(margin={'r':0, 't':0, 'l':0, 'b':0})
```
